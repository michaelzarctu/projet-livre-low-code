```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
%matplotlib inline
```

```python
url ="https://github.com/maabel0712/low-code-ai/blob/8a6d388f05b8cfe1eff9b872faa4f8529032bc62/Chapter4/advertising_2023.csv?raw=true"
advertising_df=pd.read_csv(url, index_col=0)
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(advertising_df.corr(),annot=True,vmin=0,vmax=1,cmap='viridis')
```

```python
advertising_df.plot(kind='scatter', x=['digital'], y='sales')
```

```python
plt.figure(figsize=(18, 18))

for i, col in enumerate(advertising_df.columns[0:13]):
  plt.subplot(5, 2, i+1) # deux figures par ligne
  x = advertising_df[col] # axe x
  y = advertising_df['sales'] #axe y
  plt.plot(x, y, 'o')

  #crer une ligne de rgression
  plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1)) (np.unique(x)), color='red')
  plt.xlabel(col) # tiquette x
  plt.ylabel('sales') # tiquette y
```
